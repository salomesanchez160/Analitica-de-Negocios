# ***CASO DE ESTUDIO***

En este documento se desarrolla y analiza un modelo Naive Bayes para la Detección de Lavado de Activos
en una entidad financiera utilizando las variables socioeconómicas y de comportamiento financiero de
los clientes. De acuerdo con lo anterior, las variables consideradas son:

***Edad:*** Es la edad que posee el cliente al momento del análisis.
***Ingresos:*** Ingresos declarados por el cliente en dólares estadounidenses (USD).
***Gastos:*** Son los gastos mensuales del cliente en USD.
***Número de Tarjetas de Crédito:*** Cantidad de tarjetas de crédito activas que posee el cliente.
***Monto Transado en Tarjetas:*** Monto total en USD que el cliente ha transado con sus tarjetas de crédito.
***Porcentaje de Crecimiento Patrimonial:*** Indica el porcentaje de crecimiento del patrimonio del cliente.

0. Se procede con la carga de librerías de trabajo

```python
import numpy as np #Librería numérica por excelencia
import pandas as pd #Librería para la comunicación con archivos de excel

#Librerías especificas
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1. Se cargan los datos de trabajo de la base de datos de Lavado de Activos

```python
nxl= "/content/2. LavadoActivos.xlsx"
XDB= pd.read_excel(nxl, sheet_name=0)
XDB.head (100) #Es para mostrar la base de datos cargada

#Seleccionamos variables de trabajo
XD=XDB[['edad','ingresos_usd','gastos_usd','num_tarjetas_credito','monto_transado_tarjetas_usd','porcentaje_crecimiento_patrimonio']] #Variables de trabajo
XD.head(10)
yd=XDB[['lavado_activos']]
yd.head()
```

2. Implementamos Modelo Naive Bayes

```python
mnb= GaussianNB()
mnb.fit(XD, yd) #Ajustar variables entrada - salida

#Mostramos las medias de las variables
u=mnb.theta_
sigma= mnb.var_; sigma=np.sqrt(sigma)
print("'edad','ingresos_usd','gastos_usd','num_tarjetas_credito','monto_transado_tarjetas_usd','porcentaje_crecimiento_patrimonio'")
print(u)
print("Las desviaciones son:")
print(sigma)
```

# ***ANÁLISIS DE RESULTADOS***

De los resultados arrojados por el modelo, podemos observar que la categoría de clientes con Lavado de
Activos presenta características distintivas en comparación con los clientes sin esta práctica ilícita.
Es importante mencionar que los clientes involucrados en lavado de activos tienden a mostrar patrones
de comportamiento financiero diferentes, especialmente en el crecimiento patrimonial y los montos
transados. Podemos decir igualmente, que existen diferencias significativas en el número de tarjetas
de crédito y en la relación entre ingresos y gastos, lo que indica señales de alerta para la detección
de operaciones sospechosas.

3. Se procede con la evaluación del modelo. Para la evaluación de este tipo de modelo se utiliza la matriz de confusión.

```python
ydp= mnb.predict(XD) #Esto es lo que el modelo aprende - ydp de pronóstico
cm=confusion_matrix(yd,ydp)
print(cm)

#Se determinan las métricas de la matriz de confusión
VN=cm[0,0]; FP=cm[0,1]; FN=cm[1,0]; VP=cm[1,1]; TDatos=len(XDB)

#1. Exactitud: Funcionamiento general del modelo
Ex= (VP+VN)/TDatos
print("Exactitud: ", Ex)

#2. Tasa de error: %Fallos del Modelo
TEr= (FP+FN)/TDatos
print("Tasa de Error: ", TEr)

#3. Sensibilidad: Como se comportó con respecto a los positivos solamente
Se= VP/(VP+FN)
print("Sensibilidad: ", Se)

#4. Especificidad: Como se comporta pronosticando negativos
Es= VN/(VN+FP)
print("Especificidad: ", Es)

#5. Precisión: Es una versión de como se comporta el modelo frente a los positivos solamente
Pr= VP/(VP+FP)
print("Precisión: ", Pr)

#6. Predicción Negativa: Como funciona el modelo pronosticando clientes sin lavado de activos
PrN= VN/(VN+FN)
print("Predicción Negativa: ", PrN)
```

# ***ANÁLISIS DE RESULTADOS***

De acuerdo con los resultados, podemos observar de manera general, que el modelo alcanzó una exactitud
significativa en promedio, lo que indica el buen comportamiento del modelo frente a la clasificación de
clientes según su potencial involucramiento en lavado de activos.

La precisión del modelo indica el buen funcionamiento frente a la identificación de casos positivos de
